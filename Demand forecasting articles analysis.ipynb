```python
import requests
import collections
import matplotlib.pyplot as plt
```

```python
key_words = []
no_key_words_articles_num = 0

for i in range(1, 11):
    URL = "https://doaj.org/api/search/articles/demand%20forecasting?page={}&pageSize=100".format(str(i))
    result = requests.get(URL).json()
    for response in result['results']:
        if 'keywords' in response['bibjson'].keys():
            key_words.append(response['bibjson']['keywords'])
        else:
            no_key_words_articles_num += 1
```

```python
freq = collections.Counter([x for sublist in key_words for x in sublist])

top10 = dict(freq.most_common(10))
print(top10)
```

```python
plt.xticks(rotation=60)
plt.bar(top10.keys(),top10.values(), width=1, edgecolor='black', align='edge')
plt.show()
print(no_key_words_articles_num)
```
